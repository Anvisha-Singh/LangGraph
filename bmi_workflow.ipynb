```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
class BMIState(TypedDict):
    weight: float
    height: float   
    bmi: float
    categorty: str
```

```python
def calculate_bmi_function(state: BMIState) -> BMIState:
    weight = state['weight']
    height = state['height']
    bmi = weight / (height ** 2)
    state['bmi'] = bmi
    return state

def labelbmi_function(state: BMIState) -> BMIState:
    bmi = state['bmi']
    if bmi < 18.5:
        state['categorty'] = 'Underweight'
    elif 18.5 <= bmi < 25:
        state['categorty'] = 'Normal weight'
    elif 25 <= bmi < 30:
        state['categorty'] = 'Overweight'
    else:
        state['categorty'] = 'Obesity'
    return state
    
```

```python
graph=StateGraph(BMIState)
# adding nodes to graph
graph.add_node("calculate_bmi", calculate_bmi_function)
graph.add_node("label_bmi_function", labelbmi_function)


# adding edges to graoh 
graph.add_edge(START, "calculate_bmi")

graph.add_edge("calculate_bmi", "label_bmi_function")

graph.add_edge("label_bmi_function", END)

# compile graph and run it
workflow=graph.compile()

# execute the graph and get the result
workflow.invoke({"weight": 70, "height": 1.75})
```
